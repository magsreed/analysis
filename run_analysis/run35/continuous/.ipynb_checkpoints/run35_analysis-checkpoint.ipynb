Overview
-----

This **draft** notebook contains some early analysis for the longer data set we took near the end of run 35, which also contained didv. New users, **please refer to the detprocess demo for how to set up the analyzer object, add cuts and features etc. The Semiautocut demo will also be useful here.** 

The YAML file and filter file this set was processed with are uploaded in the github processing folder. There you can look at the trace length, coincident window, low chi2 cutoff freq etc. The template that we processed with is generated and shown in the didv2 notebook. We selected the template generated from the true fit with full dIdP response, as it is a data driven template. 

I note that this is a **draft** version. Cuts especially will be updated and better documented. This is mostly for analysis constributors (like Yen Yung) to get a read on the overall flow and look at some graphs early on. I generally follow the same structure when I set up a analysis and here is that breakdown:

Quality Cuts
----
1. Baseline
    1. Baseline in time
    2. Baseline power percentile gaussian distribution 
2. Slope (not always, sometime baseline is good enough I skip this)
3. Chi2
    1. Chi2 in time
    2. Chi2 in energy (if appropriate/if we care about higher energies)
4. Low chi2
    1. Lowchi2 randoms percentile gaussian distribution
    2. Lowchi2 at low energy (do not cut lowchi2 at higher energies!)

Sanity Check
------
At this point we should sanity check our cuts to make sure we aren't throwing too much, or too little. 

1. Efficiency check
    1. True event/rate efficiency 
    2. Randoms event/rate efficiency (both of these are not in this notebook but should be done always)
2. OF Amp 1 v OF Amp 2 AND/OR Integrated Energy v OF Energy/Amp (depends on two channel or otherwise)
    1. Just a note: sometimes events or populations will appear clearly at this stage that weren't visible before. Things like glitches, or something else. OF Amp v OF Amp is the graph we look at when we do singles analysis usually, but other things can show up. 
3. Actual trace viewing
    1. Isolate a population or energy band to check if it is "real" or something we want to toss

Onwards....
-----
This is the "actual" analysis, and will depend on your analysis goals and device. Some examples of branches of possible work flows in the analysis tree are:
1. Singles/shared analysis 
2. Energy distribution / scintillation analysis
3. IR Study 

For this notebook we should actually be doing 2 of these. A singles or shared analysis for the IrPt, and an IR study (which was already completed at this time). The sapphire analysis could also be expanded on, possibly looking at energy distributions or integrated rates etc. 

All in all analysis changes or can change each time, especially with new or different devices and setups. Strange event populations can pop up, and we might need to regenerate templates, or do a deltachi2 analysis (more on that eventually). It is important to keep in mind a big goal or big picture however! If you let yourself, you can do analysis on the same data set forever! And with that, the rough notebook.

# Setup

```python
fs = 1.25e6 #sample_rate [hz]
tracelength = 2e-2 #[s]
chi2mean = fs*tracelength #the chi2 mean or the variance is equal to the number of dof per trace
# a good chi2 fit should be about the nbins
delf = fs/chi2mean
lowfreqmean = (50000/delf)*2 #50000 is the chosen cutoff frequency we used when we set up the YAML
sigchi = np.sqrt(2*lowfreqmean) #1 sigma for the low chi2

channels = ['CPDv21Al2O3','IrPtTESSmallV2', 'IrPtTESBigV2']
```

```python
# adding some regularly used or needed things in the dataframe
# I like to add time, since time since run start is more intuitive than the weird string of numbers that
# All python/computing stuff uses to track time
myanalyzer.add_feature((myanalyzer.df.event_time - myanalyzer.df.event_time.min())/60, 'timesincestart',overwrite=True)
Randcut = myanalyzer.df.trigger_type==3.0
myanalyzer.register_cut(Randcut, 'Randoms',overwrite=True) #Adding a feature which is only randoms
Trigcut = myanalyzer.df.trigger_type==4.0
myanalyzer.register_cut(Trigcut, 'Trigger',overwrite=True)
```

```python
# Here I am just reading in metadata in order to get some detector settings
# I will use this to calculate the baseline power distribution, and possibly to do some other 
# Calculations later.

# data path 
data_path = '/sdata2/runs/run35/raw/continuous_I2_D20231122_T215934' #long continuous set 
didv_bor_series = 'I2_D20231122_T215959'
file_list = glob(f'{data_path}/*{didv_bor_series}_*.hdf5')
channels = ['CPDv21Al2O3','IrPtTESSmallV2','IrPtTESBigV2']
channels_rp = [0.00365,0.00265,0.00255] #these we get from the run35 performance doc on google drive
detector_settings = h5.get_detector_config(file_name=file_list[0])
for ichan,chan in enumerate(channels):
    if ichan==1:
        rp_small = channels_rp[ichan]
        rshunt_small = float(detector_settings[chan]['shunt_resistance'])
        sg_freq_small = float(detector_settings[chan]['signal_gen_frequency'])
        sg_current_small = float(detector_settings[chan]['signal_gen_current'])
        tes_bias_small = float(detector_settings[chan]['tes_bias'])
        voltage_bias_small = tes_bias_small*rshunt_small
        p0_small = 7.68*1e-15 #watts #p0 and r_tes are from didv  @ 362.75 mOhms = 7.68 fWatts
        r_tes_small = 362.75*1e-3 #Ohms
        i_tes_small = voltage_bias_small/(r_tes_small + rp_small + rshunt_small)
    if ichan==2:
        rp_big = channels_rp[ichan]
        rshunt_big = float(detector_settings[chan]['shunt_resistance'])
        sg_freq_big = float(detector_settings[chan]['signal_gen_frequency'])
        sg_current_big = float(detector_settings[chan]['signal_gen_current'])
        tes_bias_big = float(detector_settings[chan]['tes_bias'])
        voltage_bias_big = tes_bias_big*rshunt_big
        p0_big = 196.14*1e-15 #watts #p0 and r_tes are from didv @ 301.48 mOhms = 196.14 fWatts
        r_tes_big = 301.48*1e-3 #Ohms
        i_tes_big = voltage_bias_big/(r_tes_big + rp_big + rshunt_big)

    if ichan==0:
        rp_sap = channels_rp[ichan]
        rshunt_sap = float(detector_settings[chan]['shunt_resistance'])
        sg_freq_sap = float(detector_settings[chan]['signal_gen_frequency'])
        sg_current_sap = float(detector_settings[chan]['signal_gen_current'])
        tes_bias_sap= float(detector_settings[chan]['tes_bias'])
        voltage_bias_sap = tes_bias_sap*rshunt_sap
        p0_sap = 363.15*1e-15 #watts #p0 and r_tes are from didv  @ 52.98 mOhms = 363.15 fWatts
        r_tes_sap = 52.98*1e-3 #Ohms
        i_tes_sap = voltage_bias_sap/(r_tes_sap + rp_sap + rshunt_sap)
```

## Functions

```python
def gauss(x, A, mu, sigma): #def of gaussian
    return (A * np.exp(-1.0 * (x - mu)**2 / (2 * sigma**2)))
```

### Gaussian dist

```python
# checking baseline gaussian distribution for tail location to confirm percentage choice
# looking at scale in fractional change to bias power
# I also use the astropy units package to keep track of units for me 

#-1/i0[A]*(Rl[Ohms]-R0[Ohms]
i0 = i_tes_small*U.A
rl = (rp_small+rshunt_small)*U.Ohm #Rl = Rp + Rsh
r0 = r_tes_small*U.Ohm
didp0 = 1/(i0*(rl-r0)) #units of 1/Amp Ohms or 1/V, current responsivity 
p0 = p0_small*U.W
bloffset = 2.89e-8 # this is just from eyeballing the baseline offset 
print(didp0)
didp0 = didp0.value
p0 = p0.value
```

```python
# here I just grab 400 s of randoms baseline in time to look at power variation, and split into bins
# to plot a gaussian
# I usually just eyeball where the gaussian starts to tail and cut on that, changing the percentile by hand
start_time=(myanalyzer.get_values('event_time',cut='Randoms'))[0] #
event_time_values = myanalyzer.df.event_time
ctime_temp_plotting = (event_time_values>start_time+400) & (event_time_values<start_time+800)
#ctime_temp_plotting = (vx.from_arrays(x=ctime_temp_plotting))
myanalyzer.register_cut(ctime_temp_plotting, 'bl_percentile', overwrite=True)
cut_value_plotting=np.percentile(myanalyzer.get_values('baseline_IrPtTESSmallV2', cut='bl_percentile'),95)
fig, ax = myanalyzer.hist(
    (myanalyzer.df['baseline_IrPtTESSmallV2']+bloffset)/-didp0/p0,
    cuts = [ctime_temp_plotting],
    shape = 200,
    title='IrPtsmall Baseline: Randoms',
    xlabel = '400s of baseline (fractional change compared to bias power)',
    labels = ['Baseline']
)
ax.set_yscale('linear')
ax.axvline((cut_value_plotting+bloffset)/-didp0/p0,color='red', label='95%')
ax.tick_params(which="both", direction="in",right=True, top=True)
ax.grid(linestyle="dashed")
ax.set_xlim(-0.008,0.02)
ax.legend()
print((cut_value_plotting+bloffset)/-didp0/p0)
print(didp0)

'''
fig.savefig(
    basedir+'/small_bldist_nc.png',
    dpi=500,
    facecolor='none',
)
#''';
```
