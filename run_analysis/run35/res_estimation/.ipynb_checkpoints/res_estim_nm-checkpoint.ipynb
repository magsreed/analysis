# Functions

```python
# Functions
def fold(template, fs):
    '''
    Uses numpy.fft packages to fft and fold templates. Returns both the folded and unfolded versions.
    --------------
    template [s]: 
        template in seconds
    fs [Hz]:
        sample rate in Hz    
    returns:
    ---------------
    template_f_folded: 
        Normalized folded template in frequency space.
    template_f_notfolded:
        Normalized non folded template in frequency space.
    f_folded:
        Folded frequencies. 
    f_notfolded:
        Non folded frequencies.
    '''
    #This folding done below is the same way we fold our psd's
    #see: 
    #https://github.com/spice-herald/QETpy/blob/eece3369b08bc22e924d6492df3ec1280d1fd6b5/qetpy/core/_noise.py#L72C5-L72C5
    
    template_f_folded = np.abs((rfft(template)))**2 # take the real part of the fft and its abs val and square it
    # then we fold it:
    # multiply the necessary frequencies by two (zeroth frequency should be the same, as
    # should the last frequency when x.shape[-1] is odd)
    template_f_folded[1:template.shape[-1]//2+1 - (template.shape[-1]+1)%2] *= 2.0
    f_folded = rfftfreq(len(template),d=1.0/fs)
    
    template_f_notfolded = np.abs((fft(template)))**2
    f_notfold = fftfreq(len(template), d=1.0/fs)
    
    zero_freq_folded = template_f_folded[0] 
    template_f_folded/=zero_freq_folded
    
    zero_freq_notfolded = template_f_notfolded[0]
    template_f_notfolded/=zero_freq_notfolded
    
    return(template_f_folded, f_folded, template_f_notfolded, f_notfold)
    
```

```python
def get_template_from_dpdi(phonon_pulse, dpdi):
    i_analytic = ifft(fft(phonon_pulse)/dpdi)
    i_analytic -= np.mean(i_analytic[:int(5e-3*fs)])
    i_analytic /= (min(i_analytic))
    i_analytic = np.real(i_analytic)
    return i_analytic
```

```python
def fold_fromf(template, freqs, fs):
    '''
    Uses numpy.fft packages to fft and fold templates. Returns both the folded and unfolded versions.
    Works for if you want the template at some specific frequency range.
    --------------
    template [s]: 
        template in seconds
    freqs [Hz]:
        frequencies in Hz
    fs [Hz]:
        sample rate in Hz    
    returns:
    ---------------
    template_f_norm: 
        Normalized folded template in frequency space.
    '''
    #this is how we fold our templates in the OF base class
    
    n = freqs.shape[0]
    df = fs/n
    
    template_f = np.abs(rfft(template))/n/df
    #template_f = template_f**2
    #template_f *= 2.0
    
    zero_freq = template_f[0]
    template_f_norm = template_f/zero_freq
    
    return(template_f_norm)
```

```python
#Functions
def get_energy_res(i_vals, dpdi, freqs, collection_eff=1, tau_collect=0, phonon_pulse=None):
    '''
    Calculates energy resolution. Default is dirac delta energy deposition. Pass a folded, normalized phonon pulse in frequency space
    to get a phonon pulse energy resolution. 
    ---------
    i_vals[A]:
        current from the psd in units of Amperes.
    dpdi[Volts/sqrt[Hz]]:
        dPdI in volts per square root Hz evaluated at the frequencies passed to the dPdI function.
        The frequencies used should match freqs.
    freqs[Hz]:
        The frequencies at which to calculate/integrate the energy resolution. 
        Should match the frequencies within dpdi.
    collection_eff:
        Efficiency (percentage) of phonon collection within the detector. Default is 1.
    tau_collect[s]:
        Collection fall time in units of seconds. Default is 0.
    phonon_pulse[f]:
        A folded, normalized phonon pulse template in frequency space. Should be the same length as 
        freqs and dPdI. Default is None. 
    ---------
    returns:
        energy_res_dict: dict
            A dictionary containing energy resolutions in units of eV.
        Sp[W^2/Hz]: array
            Power noise in units of W^2/Hz
        
    '''
    energy_res_dict = {}
    #Notice that we multiple by dpdi here instead of dividing by didp (as in the energy res calc code) 
    Sp = np.sqrt(i_vals)*np.abs(dpdi) #W/sqrt[Hz]
    Sp = Sp**2 #W^2/Hz
    
    omega = 2*np.pi*freqs
    if phonon_pulse is not None:
        integrand = 2*np.abs(phonon_pulse)**2/(np.pi*Sp)
        energy_res_phonon =  np.sqrt(1/np.trapz(integrand, x = omega))/constants.e
        energy_res_dict['energy_res_phonon'] = energy_res_phonon
        
    single_pole = collection_eff/(1.+ omega*tau_collect*1j)
    integrand = 2*np.abs(single_pole)**2/(np.pi*Sp)
    energy_res = np.sqrt(1/np.trapz(integrand, x = omega))/constants.e
    energy_res_dict['energy_res'] = energy_res
    
    return(energy_res_dict, Sp)
```

# Generate a phonon template

## 1PC

### Create template in power domain

```python
#t_arr_fromf = np.arange(0, (len(f_1pc)/fs)*2, 1/fs)
t_arr = np.arange(0, 5e-2, 1/fs) #be sure that the length of the data that you will be processing and the 
#length of the template (t_arr) are the same! In this case we processed with 50msec traces.

p_onepole_1_fast = qp.utils.make_template_twopole(np.asarray(t_arr), A = 1.0, 
                                             tau_r=1e-6, tau_f=55e-6, t0=10e-3, fs=1.25e6)
p_onepole_1_slow = qp.utils.make_template_twopole(np.asarray(t_arr), A = 1.0, 
                                             tau_r=1e-6, tau_f=270e-6, t0=10e-3, fs=1.25e6)
p_analytic_1 = p_onepole_1_fast*0.8 + p_onepole_1_slow*0.17  

p_analytic_1fold, f_1_fold, p_analytic_1_notfold, f1_notfold = fold(p_analytic_1, fs)

plt.plot(f_1_fold, p_analytic_1fold)
plt.title('1Pc phonon template')
plt.loglog()
plt.show()
```

### Convert phonon template to current domain

```python
#1Pc
dpdi_1pc, _ = get_dPdI_with_uncertainties(f1_notfold, true_fit_res_1pc, lgcplot=False)
```

```python
template_phonon_1pc = get_template_from_dpdi(p_analytic_1, dpdi_1pc)
```

```python
tau_p = np.abs(true_fit_res_1pc['falltimes'][0])
tau_m = np.abs(true_fit_res_1pc['falltimes'][1])

didv_twop_template = make_template_twopole(t_arr, A = 1.0, t0 =10e-3, fs = fs, tau_r=tau_p, tau_f=tau_m)
```

```python
t_arr = np.arange(0, (5e-2), 1/fs)
plt.plot(t_arr, template_phonon_1pc, label = "Phonon falltime, full dIdP response")
plt.plot(t_arr, didv_twop_template, label = "Two pole Template from dIdV Poles")

plt.xlim(9.9e-3, 10.9e-3)
plt.ylabel("Current Domain Response (Arb. Units)")
plt.xlabel("Time (s)")
plt.grid()
plt.legend()
plt.title('Phonon Template, dIdP: Melange1%')
plt.show()
```

```python
t_arr= t_arr[:-1]
```

```python
ofamp_to_ev = get_energy_normalization(t_arr, template_phonon_1pc, true_fit_res_1pc, lgc_ev=True)
ofamp_to_joules = get_energy_normalization(t_arr, template_phonon_1pc, true_fit_res_1pc, lgc_ev=False)

print("OFAmp to eV conversion (eV/amps): " + str(ofamp_to_ev))
print("OFamp to Joules conversion (Joules/amps): " + str(ofamp_to_joules))
```
