### Distribution

We want to check the scale in fractional change to bias power over time. To do so we want the:

$\frac{dI}{dP}(\omega = 0)$

Change in bias power at infinite loop gain and zero frequency. 

We can write:

$\frac{dI}{dP}(0) = \frac{1}{i_0(R_l-R_0)}$

Which is positive or negative depending on the current $i_0$ through the TES. I.e. whether we were positively or negatively biased. More on this equation and its derivation can be found in Sam's thesis, where the above is equation 2.58.

```python
# checking baseline gaussian distribution for tail location to confirm percentage choice
# looking at scale in fractional change to bias power
#-1/i0[A]*(Rl[Ohms]-R0[Ohms]
i0 = .762e-6*U.A
rl = (4e-3+5e-3)*U.Ohm #Rl = Rp + Rsh
r0 = 418e-3*U.Ohm
didp0 = 1/(i0*(rl-r0)) #units of 1/Amp Ohms or 1/V, current responsivity 
p0 = 2.428e-13*U.W
bloffset = 1.416e-6 # this is just from eyeballing the baseline offset 
print(didp0)
didp0 = didp0.value
p0 = p0.value
```

## Lowchi2 v ofamp

You should choose your lowchi2, or your frequency cutoff chi2, based on the randoms at threshold and low energies. Above this you can just do a basic percentile cut, but should also keep in mind that lowchi2 is generally most useful at threshold and low energies. Above this we should just look at the usual chi2. You can see this in how quickly (in amplitude or energy) the lowchi2 becomes bad compared to the usual chi2.

### Distribution 

Choosing the randoms lowchi2 cut based on the distribution of randoms, which will be gaussian. Our cut will be where the distribution starts to tail. 

```python
randomdata = IrPtTESSmall.get_values('lowchi2_of1x1_constrained_IrPtTESSmall',cut='Randoms')
```

```python
def gauss(x, A, mu, sigma): #def of gaussian
    return (A * np.exp(-1.0 * (x - mu)**2 / (2 * sigma**2)))
```

```python
bins, edges = np.histogram(randomdata,bins = 500,range = (0,8000)) #do the histogramming of the actual random data
bin_widths = np.diff(edges)
```

```python
popt, pcov = curve_fit(gauss, xdata=edges[2:], ydata=bins[1:], p0=[10000, 2000, 100]) #fit the hist with a gaussian. p0 are just the guesses for amp,mean,sigma 
print(popt)
xspace = np.linspace(1700,2500 , 1000) #for plotting
```

```python
plt.step(edges[1:],bins, where='mid') #plot the hist and fit together 
plt.plot(xspace, gauss(xspace, *popt), color='darkorange', linewidth=2.5, label=r'Fitted function')
plt.yscale('log')
plt.ylim(10e0,10e4)
plt.xlim(1000,3000)
plt.title(f'Randoms low chi2: Small IrPtTES $\\sigma=64$')
plt.ylabel('Counts')
plt.xlabel('Low Frequency Chi2')
plt.vlines(2200,ymin=0,ymax=10e5,colors='r', label='cut')
plt.grid(linestyle="dashed")
plt.legend()

'''
plt.savefig(
    basedir+'/small_randlowchi_dist.png',
    dpi=500,
    facecolor='none',
)
#''';
```

One sigma as above is $\sigma = 64$, and it looks like we want to cut that tail off, which is at about 2200 chi2 low freq. Meaning we want to cut $x\sigma = 200$ where $\sigma = 64$

Thus $x=3.1$. This can be translated to a P value and becomes $P = 0.198$ or about 20%. So we will want to cut 20% of our randoms, and allow 80% of data to pass. 

Please note that lowfreqchi2 cuts based on randoms are only useful in the low amplitude region.

# Singles

Singles ranges are defined as follows: 
Below 5 sigma of the energy threshold of the other device you are comparing with, and above energy threshold of the device whose singles you are examining. For example, if we are looking at the small device:


If the small device has energy resolution of $\sigma_{res} = 75$meV, triggering at 4$\sigma$ means for the small device we have an expected energy threshold of $4*(\sigma=75meV = 0.3 eV)$ 

If the big device has energy resolution of $\sigma_{res} = 375$meV, triggering at 4$\sigma$ means for the small device we have an expected energy threshold of $4*(\sigma=375meV = 1.5 eV)$

The singles on the small device will be:

above $5*(\sigma=75meV) = 0.375 eV$ and

below $5*(\sigma=375meV) = 1.875 eV$

And vice versa for the big device.

```python
def calc_rate_e_depend(e_recon, exp_time, efficiency):
    sorted_e_recon = np.sort(e_recon)
    sorted_sum = np.zeros(len(sorted_e_recon))
    for ii, e in enumerate(sorted_e_recon):
        sorted_sum[ii] = len(sorted_e_recon[ii:]) / exp_time
        
    return interpolate.interp1d(
        sorted_e_recon,
        sorted_sum/efficiency,
        bounds_error=False,
        fill_value=(sorted_sum[0], sorted_sum[-1]),
    )
```

```python
def calc_diff_rate(rate_func, energies):
    drde = []
    int_rate = np.array([rate_func(e) for e in energies])
    for ii in range(len(energies)-1):
        drde_local = (int_rate[ii+1]-int_rate[ii]) / (energies[ii+1]-energies[ii])
        drde.append(drde_local)
    return drde
```
